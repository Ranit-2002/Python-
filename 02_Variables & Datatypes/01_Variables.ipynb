# Variables:
In programming, a variable is a symbolic name or identifier that refers to a value or data stored in the computer's memory. Variables are used to store and manipulate data within a program. In Python, variables are dynamically typed, meaning you don't need to explicitly declare the type of a variable before assigning a value to it.




## Variable Naming Rules:

* Variable names can contain letters (both uppercase and lowercase), digits, and underscores (_).
* Variable names must start with a letter or an underscore (_).
* Variable names cannot start with a digit.
* Variable names are case-sensitive (my_variable is different from My_Variable).
* Variable names should be descriptive and meaningful to convey the purpose of the variable.



**Here's how variables work in Python:**

```python
# Assigning variables:

item1 = "Ballons" # Assigning a value to a variable
print(item1)

item2 = "Party hats" # Assigning another value to a variable
print(item2)

num_of_items = 2 # Assigning floating-point value to a variable
print("Total Number of Item : ", num_of_items)
```

```python
# Assigning Multiple Variable in one line
x, y = 2, 4
print(x)     # here x = 2
print(y)     # here y = 4
```
