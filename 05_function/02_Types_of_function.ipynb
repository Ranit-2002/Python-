# **Types of Function :**

Functions in Python can be categorized into several types based on their characteristics and purposes. Here are common types of functions:

* **Built-in Functions :**
These functions are part of the Python standard library and are available for use without the need for explicit definition. Examples include len(), print(), and type().


```python
word = "Hello"
print("Output :", len(word))
```

* **User-Defined Functions :**
Functions created by the programmer to perform specific tasks. They are defined using the **def** keyword.

```python
def greet():
    name = input("Enter Name :")
    print("Hello, " + name + "!")
```

```python
greet()
```

* **Recursive Functions :**
Functions that call themselves, either directly or indirectly. Recursive functions are useful for solving problems that can be broken down into smaller, similar sub-problems.

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n-1)
```

```python
factorial(0)
```

* **Anonymous Functions (Lambda Functions) :**
Small, one-line functions defined without using the def keyword. They are created using the lambda keyword.

```python
square = lambda x: x ** 2
square(2)
```
